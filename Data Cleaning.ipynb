## Citi Bike JC Data Cleaning

```python
import pandas as pd
```

```python
# Loading CSV
jan_citi_data = "data/JC-201901-citibike-tripdata.csv"

# Reading CSV into a DF
jan_df = pd.read_csv(jan_citi_data)
jan_df.dropna(inplace = True)

# Splitting the 'starttime' column to have access to more data
new = jan_df["starttime"].str.split(" ", n = 1, expand = True) 
new2 = jan_df["stoptime"].str.split(" ", n = 1, expand = True) 

# Creating new columns and assigning those columns data
jan_df["start date"]= new[0] 
jan_df["end date"]= new2[0]
jan_df["start time"]= new[1] 
jan_df["end time"]= new2[1]

# Dropping the previous column
jan_df.drop(columns =["starttime"], inplace = True)
jan_df.drop(columns =["stoptime"], inplace = True)

jan_df.to_csv('data/cleaned/jan_JC_data.csv')
jan_df.head()
```
